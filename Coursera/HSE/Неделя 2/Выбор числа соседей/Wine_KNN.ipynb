```python
import pandas as pd 
```

```python
columns = [
    "Class",
    "Alcohol",
    "Malic acid",
    "Ash",
    "Alcalinity of ash",
    "Magnesium",
    "Total phenols",
    "Flavanoids",
    "Nonflavanoid phenols",
    "Proanthocyanins",
    "Color intensity",
    "Hue",
    "OD280/OD315 of diluted wines",
    "Proline",
]
data = pd.read_csv('wine.data', names = columns, index_col = False)
```

```python
data
```

```python
y = data['Class']
X = data
del X['Class']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold, cross_val_score
```

```python
fold = KFold(shuffle = True, random_state = 42)
```

```python
scores = []
for k in range(1,51):
    clf = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(clf, X, y, cv = fold, scoring = 'accuracy').mean()
    scores.append(score)
```

```python
scores
```

```python
best_score = max(scores)
```

```python
best_k = scores.index(best_score) + 1
```

```python
print(best_k, best_score)
```

```python
from sklearn.preprocessing import scale
```

```python
X_scaled = scale(X)
```

```python
scores_scaled = []
for k in range(1,51):
    clf = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(clf, X_scaled, y, cv = fold, scoring = 'accuracy').mean()
    scores_scaled.append(score)
```

```python
best_score_scaled = max(scores_scaled)
best_k_scaled = scores_scaled.index(best_score_scaled) + 1
print(best_k_scaled, best_score_scaled)
```
